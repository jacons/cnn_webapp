# Plot results of the experiments

This notebook plots the results of the experiments
It reads the results from JSON files and generates plots for the training and validation loss, as well as the F1 score for different ResNet architectures (ResNet18 and ResNet34) with and without pretrained weights, and with different numbers of frozen layers.

```python
import pandas as pd
```

```python
r18_fre0 = pd.read_json('results/r18_fre0.json')
r18_fre0.index.rename('epoch', inplace=True)
r18_fre0.reset_index(inplace=True)
r18_fre0["epoch"] += 1
r18_fre0["name"] = "r18_fre0"

r18_pre_fre0 = pd.read_json('results/r18_pretrained_fre0.json')
r18_pre_fre0.index.rename('epoch', inplace=True)
r18_pre_fre0.reset_index(inplace=True)
r18_pre_fre0["epoch"] += 1
r18_pre_fre0["name"] = "r18_pre_fre0"

r18_pre_fre6 = pd.read_json('results/r18_pretrained_fre6.json')
r18_pre_fre6.index.rename('epoch', inplace=True)
r18_pre_fre6.reset_index(inplace=True)
r18_pre_fre6["epoch"] += 1
r18_pre_fre6["name"] = "r18_pre_fre6"

r18_pre_fre8 = pd.read_json('results/r18_pretrained_fre8.json')
r18_pre_fre8.index.rename('epoch', inplace=True)
r18_pre_fre8.reset_index(inplace=True)
r18_pre_fre8["epoch"] += 1
r18_pre_fre8["name"] = "r18_pre_fre8"

r34_fre0 = pd.read_json('results/r34_fre0.json')
r34_fre0.index.rename('epoch', inplace=True)
r34_fre0.reset_index(inplace=True)
r34_fre0["epoch"] += 1
r34_fre0["name"] = "r34_fre0"

r34_pre_fre0 = pd.read_json('results/r34_pretrained_fre0.json')
r34_pre_fre0.index.rename('epoch', inplace=True)
r34_pre_fre0.reset_index(inplace=True)
r34_pre_fre0["epoch"] += 1
r34_pre_fre0["name"] = "r34_pre_fre0"

r34_pre_fre6 = pd.read_json('results/r34_pretrained_fre6.json')
r34_pre_fre6.index.rename('epoch', inplace=True)
r34_pre_fre6.reset_index(inplace=True)
r34_pre_fre6["epoch"] += 1
r34_pre_fre6["name"] = "r34_pre_fre6"

r34_pre_fre8 = pd.read_json('results/r34_pretrained_fre8.json')
r34_pre_fre8.index.rename('epoch', inplace=True)
r34_pre_fre8.reset_index(inplace=True)
r34_pre_fre8["epoch"] += 1
r34_pre_fre8["name"] = "r34_pre_fre8"
```

```python
print(pd.concat([
    r18_fre0.tail(1),
    r18_pre_fre0.tail(1),
    r18_pre_fre6.tail(1),
    r18_pre_fre8.tail(1),
    r34_fre0.tail(1),
    r34_pre_fre0.tail(1),
    r34_pre_fre6.tail(1),
    r34_pre_fre8.tail(1),
], axis=0)[["name", "train_loss", "train_f1", "valid_loss", "valid_f1"]].reset_index(drop=True).to_latex())
```

```python
import matplotlib.pyplot as plt
from cycler import cycler
import numpy as np

colors = ['#000000', '#E69F00', '#56B4E9', '#009E73']
plt.rcParams['axes.prop_cycle'] = cycler(color=colors)

f, axes = plt.subplots(1, 2, figsize=(15, 4), sharey=True)

axes[0].plot(r18_fre0["epoch"], r18_fre0["valid_loss"], label="No pretrained", marker='o')
axes[0].fill_between(r18_fre0["epoch"],
                     r18_fre0["valid_loss"] + r18_fre0["valid_loss_std"],
                     r18_fre0["valid_loss"] - r18_fre0["valid_loss_std"],
                     color='#000000', alpha=0.2)

axes[0].plot(r18_pre_fre0["epoch"], r18_pre_fre0["valid_loss"], label="Pretrained", marker='s')
axes[0].fill_between(r18_pre_fre0["epoch"],
                     r18_pre_fre0["valid_loss"] + r18_pre_fre0["valid_loss_std"],
                     r18_pre_fre0["valid_loss"] - r18_pre_fre0["valid_loss_std"],
                     color='#E69F00', alpha=0.2)

axes[0].plot(r18_pre_fre6["epoch"], r18_pre_fre6["valid_loss"], label="Pretrained 6 freeze l.", marker='^')
axes[0].fill_between(r18_pre_fre6["epoch"],
                     r18_pre_fre6["valid_loss"] + r18_pre_fre6["valid_loss_std"],
                     r18_pre_fre6["valid_loss"] - r18_pre_fre6["valid_loss_std"],
                     color='#56B4E9', alpha=0.2)

axes[0].plot(r18_pre_fre8["epoch"], r18_pre_fre8["valid_loss"], label="Pretrained 8 freeze l.", marker='*')
axes[0].fill_between(r18_pre_fre8["epoch"],
                     r18_pre_fre8["valid_loss"] + r18_pre_fre8["valid_loss_std"],
                     r18_pre_fre8["valid_loss"] - r18_pre_fre8["valid_loss_std"],
                     color='#009E73', alpha=0.2)

axes[0].set_xlabel("Epochs", fontsize=18)
axes[0].set_ylabel("CE Loss", fontsize=17)
axes[0].set_xticks(np.arange(0, 16, 1))
axes[0].set_yscale("log")
axes[0].grid(True, linestyle="-", linewidth=0.5)
axes[0].tick_params(axis='both', labelsize=12)
axes[0].set_title("ResNet18", fontsize=20)

axes[1].plot(r34_fre0["epoch"], r34_fre0["valid_loss"], label="No pretrained", marker='o')
axes[1].fill_between(r34_fre0["epoch"],
                     r34_fre0["valid_loss"] + r34_fre0["valid_loss_std"],
                     r34_fre0["valid_loss"] - r34_fre0["valid_loss_std"],
                     color='#000000', alpha=0.2)

axes[1].plot(r34_pre_fre0["epoch"], r34_pre_fre0["valid_loss"], label="Pretrained", marker='s')
axes[1].fill_between(r34_pre_fre0["epoch"],
                     r34_pre_fre0["valid_loss"] + r34_pre_fre0["valid_loss_std"],
                     r34_pre_fre0["valid_loss"] - r34_pre_fre0["valid_loss_std"],
                     color='#E69F00', alpha=0.2)

axes[1].plot(r34_pre_fre6["epoch"], r34_pre_fre6["valid_loss"], label="Pretrained 6 freeze l.", marker='^')
axes[1].fill_between(r34_pre_fre6["epoch"],
                     r34_pre_fre6["valid_loss"] + r34_pre_fre6["valid_loss_std"],
                     r34_pre_fre6["valid_loss"] - r34_pre_fre6["valid_loss_std"],
                     color='#56B4E9', alpha=0.2)

axes[1].plot(r34_pre_fre8["epoch"], r34_pre_fre8["valid_loss"], label="Pretrained 8 freeze l.", marker='*')
axes[1].fill_between(r34_pre_fre8["epoch"],
                     r34_pre_fre8["valid_loss"] + r34_pre_fre8["valid_loss_std"],
                     r34_pre_fre8["valid_loss"] - r34_pre_fre8["valid_loss_std"],
                     color='#009E73', alpha=0.2)

axes[1].set_xlabel("Epochs", fontsize=18)
axes[1].set_xticks(np.arange(0, 16, 1))
axes[1].grid(True, linestyle="-", linewidth=0.5)
axes[1].tick_params(axis='both', labelsize=12)
axes[1].set_title("ResNet34", fontsize=20)

lines_labels = [ax.get_legend_handles_labels() for ax in axes]
lines, labels = zip(*lines_labels)
lines = sum(lines, [])
labels = sum(labels, [])

unique = dict()
for line, label in zip(lines, labels):
    if label not in unique:
        unique[label] = line

f.legend(unique.values(), unique.keys(), loc='upper center', bbox_to_anchor=(0.5, -0.15),
         ncol=4, fontsize=13, frameon=False)

f.tight_layout()
f.subplots_adjust(bottom=0)
plt.savefig('imgs/ce_loss.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler
import numpy as np

colors = ['#000000', '#E69F00', '#56B4E9', '#009E73']
plt.rcParams['axes.prop_cycle'] = cycler(color=colors)

f, axes = plt.subplots(1, 2, figsize=(15, 4), sharey=True)

axes[0].plot(r18_fre0["epoch"], r18_fre0["valid_f1"], label="No pretrained", marker='o')
axes[0].plot(r18_pre_fre0["epoch"], r18_pre_fre0["valid_f1"], label="Pretrained", marker='s')
axes[0].plot(r18_pre_fre6["epoch"], r18_pre_fre6["valid_f1"], label="Pretrained 6 freeze l.", marker='^')
axes[0].plot(r18_pre_fre8["epoch"], r18_pre_fre8["valid_f1"], label="Pretrained 8 freeze l.", marker='*')
axes[0].set_xlabel("Epochs", fontsize=18)
axes[0].set_ylabel("F1 Score", fontsize=17)
axes[0].set_yticks(np.arange(0, 2, 0.1))
axes[0].set_xticks(np.arange(0, 16, 1))
axes[0].grid(True, linestyle="-", linewidth=0.5)
axes[0].tick_params(axis='both', labelsize=12)
axes[0].set_title("ResNet18", fontsize=20)

axes[1].plot(r34_fre0["epoch"], r34_fre0["valid_f1"], label="No pretrained", marker='o')
axes[1].plot(r34_pre_fre0["epoch"], r34_pre_fre0["valid_f1"], label="Pretrained", marker='s')
axes[1].plot(r34_pre_fre6["epoch"], r34_pre_fre6["valid_f1"], label="Pretrained 6 freeze l.", marker='^')
axes[1].plot(r34_pre_fre8["epoch"], r34_pre_fre8["valid_f1"], label="Pretrained 8 freeze l.", marker='*')
axes[1].set_xlabel("Epochs", fontsize=18)
axes[1].set_yticks(np.arange(0, 1.2, 0.2))
axes[1].set_xticks(np.arange(0, 16, 1))
axes[1].grid(True, linestyle="-", linewidth=0.5)
axes[1].tick_params(axis='both', labelsize=12)
axes[1].set_title("ResNet34", fontsize=20)

lines_labels = [ax.get_legend_handles_labels() for ax in axes]
lines, labels = zip(*lines_labels)
lines = sum(lines, [])
labels = sum(labels, [])

unique = dict()
for line, label in zip(lines, labels):
    if label not in unique:
        unique[label] = line

f.legend(unique.values(), unique.keys(), loc='upper center', bbox_to_anchor=(0.5, -0.15),
         ncol=4, fontsize=13, frameon=False)

f.tight_layout()
f.subplots_adjust(bottom=0)
plt.savefig('imgs/f1_score.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
